# Model

 - **Model:** Gradient Boost Classifier
   
 - **Implemented with:** [CatBoostClassifier](https://catboost.ai/docs/en/concepts/python-reference_catboostclassifier)
  
 - **Loss Function:** [Logloss](https://catboost.ai/docs/en/concepts/loss-functions-classification#Logit)
  
 - **Target:** `has_seen_movie` (as proxy we use if user has rated the movie)
  
 - **Description:**  
   The model predicts the category (1 or 0) for target `has_seen_movie`.  
   The [predict_proba](https://catboost.ai/docs/en/concepts/python-reference_catboostclassifier_predict_proba) can then be used to get the probability that a user has seen a movie.  
   The ranker score is the probability a user has seen a movie.
