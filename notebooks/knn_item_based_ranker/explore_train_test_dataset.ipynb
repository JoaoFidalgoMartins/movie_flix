# Model

 - **Implemented with:** [Scikit-Learn NearestNeighbors](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.NearestNeighbors.html)
  
 - **Description:**  
   Movie to movie recommendations.  
   Given the movie features, calculates de cosine distance between movies.  
   Picks the top n closest movies.  
   To recommend movies to a user we pick the last movie the user saw and do a movie to movie recommendation.
